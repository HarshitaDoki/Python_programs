```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array(['clad', 'phy','m-1','m-2','eng','c','evpr','vdc'])
y = np.array([76,70,65,96,73,60,80,90])

plt.title("Sem_1")
plt.xlabel("Subject")
plt.ylabel("Marks")

plt.plot(x, y)

plt.grid()

plt.show()
```
